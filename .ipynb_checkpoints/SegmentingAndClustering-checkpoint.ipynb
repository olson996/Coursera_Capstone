```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
!pip install geopy 
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
import seaborn as sns
import matplotlib.pyplot as plt
import folium
import json
from sklearn import preprocessing
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
print('\nPackages installed!')
```

Now that we have the packages we need we can send out a request for the html data we need. We access the table by it's tag in <i>soup.table</i>, and read that html into a pandas dataframe.

```python
html = requests.get('http://www.laalmanac.com/economy/ec37b.php')
soup = BeautifulSoup(html.text, 'html.parser')
df = pd.read_html(str(soup.table))[0]
```

Let's clean up this data frame and name the columns. The data from the Los Angeles almanac includes each averge housing price for years 2013-2018. For this research all years prior to 2018 will be dropped from the dataframe. The price itself is in US dollar format and a regular expression is used to convert it to an integer value. 

```python
df.columns= ['Zip Code', 'Community', 'House Price', '2017', '2016', '2015', '2014', '2013']
df.drop(columns=['2017', '2016', '2015', '2014', '2013'], inplace=True)
df.dropna(inplace=True)
df['House Price'] = df['House Price'].replace('[\$,]', '', regex=True).astype(int)
df.sort_values(by=['House Price'], ascending=False, inplace=True)
df.reset_index(inplace=True)
df.drop(columns=['index'], inplace=True)
df.set_index('Zip Code', inplace=True)
df.columns = ['Community', 'House Price']
print(df.shape)
df.head()
```

We still need the coordinates to be able to find the most common venues around communities in LA. The function <i>get_lat_lon</i> takes a <b>pandas</b> dataframe as a parameter and makes a call to the <b>geopy</b> library to resolve the coordinates of each zipcode.

```python
def get_lat_lon(df):
    for i in range(0, len(df)):
        zipcode = str(df.iloc[i, :].name)
        geolocator = Nominatim(user_agent='LA_agent')
        location = geolocator.geocode(zipcode)
        df.loc[df.index[i], 'Latitude'] = location.latitude
        df.loc[df.index[i], 'Longitude'] = location.longitude
```

<b>Note:</b> The function below resolves longitiude and latitude coordinates with geolocator api and may take a couple minutes.

```python
get_lat_lon(df)
```

This function will get the nearest venues to all of our zip codes. It sends a request to Foursquares API which can be seen in the url variable below. Most of the parameters to query can be viewed above with the exception of the longitude and latitude that varies for each zip code. A new dataframe is created and returned that contains all venues around the area within a specified distance.

```python
def getNearbyVenues(zips, latitudes, longitudes, radius=500):
    venues_list=[]
    for zipcode, lat, lng in zip(zips, latitudes, longitudes):
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            zipcode, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Zip Code', 
                  'Zip Latitude', 
                  'Zip Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

Since we have all the venues in each area we can find out the frequency of each venue category for each zip code.

```python
# one hot encoding
la_onehot = pd.get_dummies(la_venues[['Venue Category']], prefix="", prefix_sep="")
# add zip code column back to dataframe
la_onehot['Zip Code'] = la_venues['Zip Code'] 
# move zip code column to the first column
fixed_columns = [la_onehot.columns[-1]] + list(la_onehot.columns[:-1])
la_onehot = la_onehot[fixed_columns]
la_grouped = la_onehot.groupby('Zip Code').mean().reset_index()
```

```python
la_grouped.head()
```

```python
la_grouped = la_grouped.join(df[["Zip Code", "Community"]].set_index("Zip Code"), on="Zip Code")
```

This function will create a new data frame of the most frequently found venue categories in an area. 

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:-1]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Zip Code', 'Community']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Zip Code'] = la_grouped['Zip Code']
neighborhoods_venues_sorted['Community'] = la_grouped['Community']

for ind in np.arange(la_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 2:] = return_most_common_venues(la_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head(5)
```

<h4>Unsupervised Learning:</h4>

This notebook uses K-Means from sklearn to cluster and label the data using the most frequent venues of an area, and the average housing price for a single family home. I performed multiple iterations of this clustering using the normalized housing price, without normalizing the housing price and without the housing price at all and only the most frequent venue categories.

The following code cell normalizes the data frame. 

```python
X = la_grouped[['House Price']].values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(X)
la_grouped['House Price'] = pd.DataFrame(x_scaled)
```

```python
la_grouped.head()
```

The type of cluster changes if you normalize the data or not, and if you include the housing price. Re-run from frequency table code cell to see different versions. 

```python
# set number of clusters
kclusters = 6

la_grouped_clustering = la_grouped.drop('Community', 1)
la_grouped_clustering = la_grouped_clustering.drop('Zip Code', 1)
#uncomment if wanting to only cluster on venues
la_grouped_clustering = la_grouped_clustering.drop('House Price', 1)
# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(la_grouped_clustering)

# check cluster labels generated for each row in the dataframe
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
# add clustering labels
# 265 vs 229
la_merged = df
# merge la_grouped with la_data to add latitude/longitude for each neighborhood
la_merged = la_merged.merge(neighborhoods_venues_sorted, on='Zip Code', how='inner')
la_merged=la_merged.drop(['level_0', 'index'], 1)
```

```python
la_merged.tail(5)
```

```python
# create map
map_clusters = folium.Map(location=[34.0522, -118.2437], zoom_start=9, tiles='Stamen Terrain')

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for community, lat, lon, poi, cluster in zip(la_merged['Community_x'], la_merged['Latitude'], la_merged['Longitude'], la_merged['Zip Code'], la_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster) + str('\n Community: \n'+community+'\n'), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
     
map_clusters.save(outfile = 'la_cluster_house_price_not_included.html')
map_clusters
```

The following function lists the number of times a category appears in the top three nearest category list in each cluster. 

You can think of this list like the 'Most common venue categories among a cluster.'

```python
def common_cluster_venues(la_merged, k):
    cluster_i = la_merged.loc[la_merged['Cluster Labels'] == k, la_merged.columns[[1] + list(range(5, la_merged.shape[1]))]]
    ser = pd.Series()
    for i in range(0, 3):
        t_series = cluster_i.iloc[:, 3+i]
        ser = pd.concat([ser, t_series])
    print(ser.value_counts()[0:3])
    print('Average Single Home Price for Cluster $ %.2f' % la_merged.loc[la_merged['Cluster Labels']==k, 'House Price'].mean())
    cluster_i = cluster_i.drop('Community_y', 1)
    cluster_i.rename(columns={'Community_x':'Community'}, inplace=True)
    return cluster_i
```

<h2><a id="results">Results & Discussion:</a> </h2>

When the data is not normalized K-Means assigns greater weight to the housing price in regard to its distance measure or similarity to other areas. In the non-normalized example above, the areas of higher housing price were clustered together, and lower ones were as well. The pattern of clustering seemed to resemble the choropleth map we viewed earlier in which the west coast and communities around Hollywood Hills and over the value had a higher mean single-family housing price. The normalized example doesnt quite cluster in to delineated groups. The normalized example does cluster the higher mean priced housing areas, it fails to cluster the lower mean priced ones.
